To perform matrix multiplication of matrix of m,k, and matrix of k,n to get each value we will have to multiply k times and add k-1 times, so for the whole matrix which consists of $m n$ elements we multiply $k m n$ times and add $(k-1) m n$ times. 

```python
#QN-2
#taking m=5,n=6,k=4
import time
import random

m=107
n=84
k=100
matr1=[]
matr2=[]
for i in range(m):
    listmatr1=[]
    for j in range(k):
        listmatr1.append(random.randint(0,9))
    matr1.append(listmatr1)
for i in range(k):
    listmatr2=[]
    for j in range(n):
        listmatr2.append(random.randint(0,9))
    matr2.append(listmatr2)

ans = [[0 for _ in range(n)] for _ in range(m)]
x=y=0
start_time=time.time()
for x in range(m):
    for y in range(n):
        ans[x][y]=0
        for z in range(k):
            ans[x][y]+=(matr1[x][z])*matr2[z][y]

print("time taken(in milliseconds) using lists- ",(time.time()-start_time)*1000)

        
```

```python

import numpy as np
start_time=time.time()
npmat1=np.array(matr1)
npmat2=np.array(matr2)
npans=np.dot(npmat1,npmat2)
start_time=time.time()
print("time taken(in milliseconds) using numpy methods- ",(time.time()-start_time)*1000)
```

From the above results, using numpy to multiply matrices is faster than using lists and loops. numpy stores data in continuous manner and it is written in C language so, by using numpy the calculation are faster.

If finding the maximum number takes one pass of the array and finding second maximum takes two passes then to find median which is approximately (n/2)th maximum number, it will take n/2 passes and if each pass takes kn time(k is some constant) then to find median it takes $ k n^2 $ so, it is of time complexity O($ n^2 $)

The time complexity can be improved by first sorting the list with quicksort which has time complexity $n logn$ and to find the middle term in the list, it takes constant time so, the total time complexity can be improved to $n logn$

```python
#by sorting
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot=arr[len(arr)//2]
    left=[x for x in arr if x<pivot]
    middle=[x for x in arr if x==pivot]
    right=[x for x in arr if x>pivot]
    return quicksort(left)+middle+quicksort(right)
n=np.random.randint(1,100000)
arr=np.random.randint(1,1000,n)
st=time.time()
throughqs=quicksort(arr)
if n%2==0:
    print("median calculated after sorting- ",(throughqs[n//2 -1] + throughqs[n//2])/2)
else:
    print("median calculated after sorting- ",throughqs[n//2])
print("time taken when using 1st method ",time.time()-st)
st=time.time()
print("median calculated using np.median()- ",np.median(arr))
print("time taken when using np.median() ",(time.time()-st)*1000)
#np.median is much faster than finding median using sorting method
```

```python
def mini(arr):
    n=len(arr)
    ans=0
    for i in range(n):
        if arr[i]<arr[ans]:
            ans=i
    return ans
arr=np.array([5,3,7,4,2])
print(mini(arr))
print(np.argmin(arr))
        
```
